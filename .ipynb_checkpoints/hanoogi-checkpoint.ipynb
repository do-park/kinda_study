```python
import collections

def safe():
    cnt = 1
    dy = [-1, 1, 0, 0]
    dx = [0, 0, -1, 1]
    while queue:
        ny, nx = queue.popleft()
        for d in range(4):
            Y = ny + dy[d]
            X = nx + dx[d]
            if 0 <= Y < N and 0 <= X < N and not visited[Y][X] and maps[Y][X] != -1:
                visited[Y][X] = True
                queue.append([Y, X])
                cnt += 1
    return cnt

N = int(input())
maps = [list(map(int, input().split())) for _ in range(N)]
ans = 1
for i in range(1, 101):
    flag = False
    for y in range(N):
        for x in range(N):
            if maps[y][x] == i:
                maps[y][x] = -1
                flag = True
    if flag:
        queue = collections.deque()
        visited = [[False] * N for _ in range(N)]
        result = []
        for x in range(N):
            for y in range(N):
                if not visited[y][x]:
                    visited[y][x] = True
                    if maps[y][x] != -1:
                        queue.append([y, x])
                        result.append(safe())
        if len(result) >= ans:
            ans = len(result)
print(ans)
```

```python
import collections

def warming(iceberg):
    temp = [[0] * M for _ in range(N)]
    for n in range(N):
        for m in range(M):
            if iceberg[n][m] != 0:
                cnt = 0
                for d in range(4):
                    Y = dy[d] + n
                    X = dx[d] + m
                    if 0 <= X < M and 0 <= Y < N and iceberg[Y][X] == 0:
                        cnt += 1
                temp[n][m] = max(iceberg[n][m] - cnt, 0)
    return temp


def dfs(iceberg, visited):
    while Q:
        y, x = Q.popleft()
        for d in range(4):
            Y = dy[d] + y
            X = dx[d] + x
            if 0 <= X < M and 0 <= Y < N and iceberg[Y][X] != 0 and not visited[Y][X]:
                visited[Y][X] = True
                Q.append([Y, X])




N, M = map(int, input().split())
iceberg = [list(map(int, input().split())) for _ in range(N)]

dy = [-1, 1, 0, 0]
dx = [0, 0, -1, 1]
Q = collections.deque()
year = 0
while True:
    visited = [[False] * M for _ in range(N)]
    end = 0
    iceberg = warming(iceberg)
    year += 1
    sep = 0
    for n in range(N):
        for m in range(M):
            if iceberg[n][m] == 0:
                end += 1
            if iceberg[n][m] != 0 and not visited[n][m]:
                visited[n][m] = True
                Q.append([n, m])
                dfs(iceberg, visited)
                sep += 1
            if sep > 1: break
        if sep > 1: break
    if sep > 1:
        print(year)
        break
    if end == N * M:
        print(0)
        break
```

```python
import collections

N, M = map(int, input().split())
island = []
Q = collections.deque()
for n in range(N):
    temp = list(input())
    for m in range(M):
        if temp[m] == 'L':
            Q.append([n, m])
    island.append(temp)

subQ = collections.deque()
dy = [-1, 1, 0, 0]
dx = [0, 0, -1, 1]
cnt = 0
subcnt = 0
while Q:
    visited = [[0] * M for _ in range(N)]
    y, x = Q.popleft()
    visited[y][x] = 1
    subQ.append([y, x])
    while subQ:
        y, x = subQ.popleft()
        for d in range(4):
            Y = y + dy[d]
            X = x + dx[d]
            if 0 <= Y < N and 0 <= X < M and island[Y][X] == 'L' and not visited[Y][X]:
                visited[Y][X] = visited[y][x] + 1
                subQ.append([Y, X])
                if subcnt < visited[Y][X]:
                    subcnt = visited[Y][X]
    if cnt < subcnt:
        cnt = subcnt
print(cnt - 1)
```

```python
from collections import deque

def melt():
    loss = 0
    while Q:
        y, x = Q.popleft()
        cheese[y][x] = 0
        for d in range(4):
            ny = y + dy[d]
            nx = x + dx[d]
            if 0 <= ny < N and 0 <= nx < M and not visited[ny][nx]:
                if cheese[ny][nx]:
                    Q_temp.append([ny, nx])
                    visited[ny][nx] = True
                    loss += 1
                else:
                    Q.append([ny, nx])
                    visited[ny][nx] = True
    return loss


N, M = map(int, input().split())
cheese = []
c = 0
for n in range(N):
    temp = list(map(int, input().split()))
    for m in range(M):
        if temp[m]:
            c += 1
    cheese.append(temp)
visited = [[False] * M for _ in range(N)]
Q, Q_temp = deque(), deque()
Q.append([0, 0])
visited[0][0] = True
dy = [-1, 1, 0, 0]
dx = [0, 0, -1, 1]
time = 1
while True:
    l = melt()
    if c == l:
        print(time)
        print(c)
        break
    Q = Q_temp
    Q_temp = deque()
    c -= l
    time += 1
```

```python
# 2667 단지번호붙이기
```

```python
def safe(x, y):
    if (0 <= x < N and 0 <= y < N):
        return True
    else:
        return False


def BFS(apart, i, j, N, visited):
    if apart[i][j] == 0:
        visited.append([i, j])
        return [0, visited]

    block = []
    queue = [[i, j]]

    while queue:
        [i, j] = queue.pop()
        block.append([i, j])
        visited.append([i, j])

        if apart[i][j] == 1:
            dy = [-1, 1, 0, 0]
            dx = [0, 0, -1, 1]
            for k in range(0, 4):
                Y = dy[k] + i
                X = dx[k] + j
                if safe(X, Y) and apart[Y][X] == 1 and [Y, X] not in block and [Y, X] not in queue:
                    queue.append([Y, X])

    return [len(block), visited]

N = int(input())
apart = [list(map(int, input())) for _ in range(N)]
visited = []
result = []



for i in range(0, N):
    for j in range(0, N):
        if [i, j] not in visited:
            [size, visited] = BFS(apart, i, j, N, visited)
            if size != 0:
                result.append(size)

print(len(result))
for i in sorted(result):
    print(i)
```
